```python
import cv2
import matplotlib.pyplot as plt
import numpy as np

img = cv2.imread('lena.png') 
height, width, _ = img.shape

histgram = [0]*256
for i in range(height):
    for j in range(width):
        histgram[img[i, j, 0]] += 1

# エントロピーの算出
size = height * width
entropy = 0

for i in range(256):
    # レベルiの出現確率p
    p = histgram[i]/size
    if p == 0:
        continue
    entropy -= p*np.log2(p)

plt.imshow(img)
print('エントロピー：{}'.format(entropy))
```
