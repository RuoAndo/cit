```python
import numpy as np
import matplotlib.pyplot as plt
import random
    
value =[]
x = []
y = []

N1 = 5000
for i in range(N1):
    value.append(random.randint(50, 70))
    x.append(random.randint(0, 100))
    y.append(random.randint(0, 100))

N2 = 500
for i in range(N2):
    value.append(random.randint(70, 100))
    x.append(random.randint(0, 100))
    y.append(random.randint(0, 100))
    
plt.scatter(x, y, s=30, c=value, cmap='Blues')

plt.colorbar()
```
