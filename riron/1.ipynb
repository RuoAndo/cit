## 1. 情報エントロピー

```python
from PIL import Image
filename = "entropy.png"
im = Image.open(filename)
im
```

```python
import numpy as np
import matplotlib.pyplot as plt

P0 = np.arange(0,1, 0.01) 
P1 = 1 - P0 

# エントロピー
H = -P0 * np.log2(P0) -P1 * np.log2(P1)

# グラフ描画設定
plt.plot(P0, H)
plt.xlabel("P0")
plt.ylabel("H")
plt.title('Entropy')

# ラベルの描画
plt.legend()
# グラフの描画実行
plt.show()
```

```python
from PIL import Image
filename = "Jensen.png"
im = Image.open(filename)
im
```

```python
import matplotlib.pyplot as plt 
import random

x=[1,2,3,4,5,6,7]
y=[2,1,4,7,4,3,2]

x = []
y1 = []
y2 = []

for i in range(500):
    x.append(random.randint(0, 100))

for i in range(500):
    y1.append(random.randint(0, 100))
    y2.append(random.randint(0, 100))

    
plt.scatter(x,y1,c="red", s=1)
plt.scatter(x,y2,c="blue", s=1)

plt.xlabel("X")
plt.ylabel("Y")
plt.title("500-500")
plt.show()
```

## 3. シャノンの第２基本定理

```python
from PIL import Image
filename = "shannon2.png"
im = Image.open(filename)
im
```

```python
import numpy as np
import matplotlib.pyplot as plt

SN= np.arange(0,50, 0.01) 
B = 1

C = B * np.log2(1+SN)

plt.plot(SN, C)
plt.xlabel("Signal-Noise Ratio")
plt.ylabel("Capacity")
plt.title('Shannon Formula 2')

plt.legend()
plt.show()
```
