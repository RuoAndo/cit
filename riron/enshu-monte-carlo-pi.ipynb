```python
# import math
import numpy as np
import matplotlib.pyplot as plt

x_all=[]
y_all=[]
pi = []
pi_mean = []
count=[]

N1 = 50
N2 = 100

xx = []
yy = []

xxx = []
yyy = []

for i in range(N1):
    count.append(i)
    for i in range(N2):
        incount = 0
        allcount = 0
        x = np.random.rand()
        y = np.random.rand()
        x_all.append(x)
        y_all.append(y)
        
        if x ** 2 + y ** 2 < 1:
            incount += 1
            xx.append(x)
            yy.append(y)
        
        else:
            xxx.append(x)
            yyy.append(y)
                  
        allcount += 1
        pi.append(4 * incount / allcount)
    pi_mean.append(np.mean(pi))
    
plt.scatter(xx,yy, color = (1,0,0))
plt.scatter(xxx,yyy, color = (0,0,1))
plt.show()

plt.plot(count,pi_mean)
plt.show()
```
