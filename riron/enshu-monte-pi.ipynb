```python
import math
import numpy as np
import matplotlib.pyplot as plt

x_all=[]
y_all=[]
pi = []
pi_mean = []
count=[]
N = 100

fig = plt.figure(figsize=(18, 6))

ax1 = fig.add_subplot(1, 2, 1)
ax2 = fig.add_subplot(1, 2, 2)

xx = []
yy = []

xxx = []
yyy = []

for i in range(100):
    count.append(i)
    for i in range(N):
        incount = 0
        allcount = 0
        x = np.random.rand()
        y = np.random.rand()
        x_all.append(x)
        y_all.append(y)
        
        if x ** 2 + y ** 2 < 1:
            incount += 1
            xx.append(x)
            yy.append(y)
        
        else:
            xxx.append(x)
            yyy.append(y)
                  
        allcount += 1
        pi.append(4 * incount / allcount)
    pi_mean.append(np.mean(pi))
    
ax1.scatter(xx,yy, color = (1,0,0))
ax1.scatter(xxx,yyy, color = (0,0,1))

ax2.plot(count,pi_mean)

fig.show()
```
