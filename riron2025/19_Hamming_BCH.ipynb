```python
import random
import matplotlib.pyplot as plt
import matplotlib

# === 日本語フォントの設定（Windows向け。macOSやLinuxはコメント参照） ===
# Windowsの場合（例：MS Gothic）
matplotlib.rcParams['font.family'] = 'MS Gothic'

# macOSの場合（例：Hiragino Sans）
# matplotlib.rcParams['font.family'] = 'Hiragino Sans'

# Linuxの場合（例：IPAexGothic。インストール必要）
# matplotlib.rcParams['font.family'] = 'IPAPGothic'

# === 実験設定 ===
trials = 1000
max_errors = 5
hamming_success = []
bch_success = []

# === シミュレーション開始 ===
for errors in range(1, max_errors + 1):
    hamming_correct = sum(1 for _ in range(trials) if errors <= 1)
    bch_correct = sum(1 for _ in range(trials) if errors <= 2)

    hamming_success.append(hamming_correct / trials)
    bch_success.append(bch_correct / trials)

# === グラフ描画 ===
x = list(range(1, max_errors + 1))
plt.figure(figsize=(10, 6))
plt.plot(x, hamming_success, marker='o', label='Hamming(7,4)')
plt.plot(x, bch_success, marker='s', label='BCH(63,51)（2ビット訂正仮定）')
plt.xlabel('誤りビット数')
plt.ylabel('訂正成功率')
plt.title('Hamming符号とBCH符号の誤り訂正性能比較（シミュレーション）')
plt.grid(True)
plt.legend()
plt.ylim(0, 1.05)
plt.xticks(x)
plt.tight_layout()
plt.show()
```
