```python
import matplotlib.pyplot as plt

# 素数 p の有限体 F_p 上での楕円曲線 y^2 = x^3 + 7
p = 103
points = []

# 楕円曲線の点を列挙
for x in range(p):
    rhs = (x ** 3 + 7) % p
    for y in range(p):
        if (y * y) % p == rhs:
            points.append((x, y))

# 無限遠点を加える（任意）
points.append(("O", "infinity"))

# 点の表示
print(f"Total number of points on E(F_{p}): {len(points)}")
for pt in points[:-1]:  # 最後の 'O' を除いて表示
    print(pt)

# プロット用に整数座標だけ取り出し
finite_points = [(x, y) for x, y in points if isinstance(x, int)]

# プロット
x_vals = [x for x, y in finite_points]
y_vals = [y for x, y in finite_points]

plt.figure(figsize=(8, 6))
plt.scatter(x_vals, y_vals, s=20)
plt.title("Elliptic Curve y^2 = x^3 + 7 over F103")
plt.xlabel("x")
plt.ylabel("y")
plt.grid(True)
plt.show()
```
