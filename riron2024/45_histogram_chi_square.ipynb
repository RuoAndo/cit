```python
import numpy as np
import matplotlib.pyplot as plt

# カイ二乗分布のパラメータ
df = 10  # 自由度

# カイ二乗分布に従う乱数を1000個生成
chi_squared_samples = np.random.chisquare(df, 1000)

# ヒストグラムの作成
plt.hist(chi_squared_samples, bins=30, edgecolor='black', alpha=0.75)
plt.title('Histogram of 1000 samples from a Chi-Squared Distribution')
plt.xlabel('Value')
plt.ylabel('Frequency')
plt.grid(True)
plt.show()
```
