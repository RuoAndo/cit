```python
import matplotlib.pyplot as plt
from collections import Counter

# 文字列の定義
string = "RRKBSNIAKTFQVZRPSBCZSHXIQRYVTDURMARKCWIYBJDNVUSWQSJH"

# 各アルファベットの出現頻度をカウント
counter = Counter(string)

# データを取得
labels, counts = zip(*counter.items())

# ヒストグラムを作成
plt.bar(labels, counts)

# グラフのタイトルとラベルを設定
plt.title('Frequency of Each Alphabet in the String')
plt.xlabel('Alphabet')
plt.ylabel('Frequency')

# グラフを表示
plt.show()
```
