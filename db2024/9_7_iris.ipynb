```python
# 必要なライブラリをインポート
from sklearn.datasets import load_iris
from sklearn.model_selection import train_test_split
from sklearn.linear_model import LogisticRegression
from sklearn.metrics import accuracy_score
import pandas as pd
import seaborn as sns
import matplotlib.pyplot as plt

# データセットのロード
iris = load_iris()

# データフレームに変換
df = pd.DataFrame(iris.data, columns=iris.feature_names)
df['target'] = iris.target
df['species'] = df['target'].apply(lambda x: iris.target_names[x])

# データの概要を表示
print("データの概要:")
print(df.head())

# データの可視化: ペアプロット
print("ペアプロットを作成中...")
sns.pairplot(df, hue='species', vars=iris.feature_names)
plt.show()

# データの可視化: 特徴量ごとの分布
print("特徴量ごとの分布を作成中...")
for feature in iris.feature_names:
    plt.figure(figsize=(6, 4))
    sns.histplot(data=df, x=feature, hue='species', kde=True, palette="bright")
    plt.title(f"{feature} の分布")
    plt.xlabel(feature)
    plt.ylabel("Count")
    plt.show()

# 特徴量間の散布図（例: petal length vs petal width）
print("特徴量間の散布図を作成中...")
plt.figure(figsize=(6, 4))
sns.scatterplot(data=df, x="petal length (cm)", y="petal width (cm)", hue="species", palette="deep")
plt.title("Petal Length vs Petal Width")
plt.xlabel("Petal Length (cm)")
plt.ylabel("Petal Width (cm)")
plt.legend(title="Species")
plt.show()

# モデルの訓練と評価
# 特徴量とラベルを定義
X = iris.data
y = iris.target

# 訓練データとテストデータに分割
X_train, X_test, y_train, y_test = train_test_split(X, y, test_size=0.3, random_state=42)

# ロジスティック回帰モデルの作成
model = LogisticRegression(max_iter=200)

# モデルの訓練
print("モデルを訓練中...")
model.fit(X_train, y_train)

# テストデータで予測
y_pred = model.predict(X_test)

# 精度の評価
accuracy = accuracy_score(y_test, y_pred)
print(f"モデルの精度: {accuracy:.2f}")

# サンプルデータの予測例
sample_data = [[5.1, 3.5, 1.4, 0.2]]  # Setosaの特徴量例
prediction = model.predict(sample_data)
predicted_species = iris.target_names[prediction[0]]
print(f"サンプルデータ {sample_data} の予測結果: {predicted_species}")
```
