```python
import yfinance as yf
import pandas as pd
import sqlite3

# -----------------------------------------------------
# ① 任天堂（7974.T）の株価を取得
# -----------------------------------------------------
df = yf.download(
    '7974.T',
    start='2018-01-01',
    end='2022-12-31'
)

# CSV に保存
df.to_csv("nintendo_2018_2022.csv")


# -----------------------------------------------------
# ② CSV を通常読み込み（yfinance は1段ヘッダ）
# -----------------------------------------------------
df = pd.read_csv("nintendo_2018_2022.csv")

# Adj Close → AdjClose の表記統一
df = df.rename(columns={"Adj Close": "AdjClose"})

# 必要な列だけ抽出
needed = ["Date", "Open", "High", "Low", "Close", "AdjClose", "Volume"]
df = df[needed]

# Date を文字列として整形
df["Date"] = df["Date"].astype(str)


# -----------------------------------------------------
# ③ SQLite に保存
# -----------------------------------------------------
conn = sqlite3.connect("nintendo.db")
df.to_sql("nintendo", conn, if_exists="replace", index=False)
conn.close()

print("SQLite nintendo.db に登録完了")
```
