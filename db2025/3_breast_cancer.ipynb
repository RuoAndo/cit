```python
import matplotlib.pyplot as plt
from sklearn.datasets import load_breast_cancer
from sklearn.decomposition import PCA

# === データ読み込み ===
data = load_breast_cancer()
X = data.data
y = data.target
features = data.feature_names

# === 1. 散布図（特徴量2つを選んで表示） ===
plt.figure(figsize=(7, 6))
plt.scatter(X[:, 0], X[:, 1], c=y, cmap='coolwarm', alpha=0.7)
plt.xlabel(features[0])  # mean radius
plt.ylabel(features[1])  # mean texture
plt.title("Scatter plot: mean radius vs mean texture")
plt.colorbar(label="Target (0=悪性, 1=良性)")
plt.show()

# === 2. PCAで2次元に次元削減して可視化 ===
pca = PCA(n_components=2)
X_pca = pca.fit_transform(X)

plt.figure(figsize=(7, 6))
plt.scatter(X_pca[:, 0], X_pca[:, 1], c=y, cmap='coolwarm', alpha=0.7)
plt.xlabel("PCA 1")
plt.ylabel("PCA 2")
plt.title("PCA projection of Breast Cancer Dataset")
plt.colorbar(label="Target (0=悪性, 1=良性)")
plt.show()
```
