```python
# ============================================================
# 教育用通信シミュレータ
# LFSR + CRC + Hamming(7,4)
# ランダム規約多項式 + 日本語表示 + plt.show() + y/N確認
# ============================================================

import random
import time
import sys
import matplotlib.pyplot as plt
from matplotlib.patches import Rectangle


AUTO_MODE = False
SHOW_PLOTS = True


PRIMITIVE_POLYNOMIALS = {
    4:  [[4, 1, 0], [4, 3, 0]],
    5:  [[5, 2, 0], [5, 4, 2, 1, 0]],
    6:  [[6, 1, 0], [6, 5, 0]],
    7:  [[7, 1, 0], [7, 3, 0]],
    8:  [[8, 4, 3, 2, 0], [8, 6, 5, 4, 0]],
    10: [[10, 3, 0], [10, 7, 0]],
    12: [[12, 6, 4, 1, 0]],
    16: [[16, 12, 3, 1, 0], [16, 5, 3, 2, 0]],
    24: [[24, 4, 3, 1, 0]],
    32: [[32, 22, 2, 1, 0]],
    48: [[48, 28, 27, 1, 0]],
    64: [[64, 4, 3, 1, 0]],
}


def wait_next():
    global AUTO_MODE

    time.sleep(random.uniform(1.0, 2.0))

    if AUTO_MODE:
        return

    ans = input("\n次へ進みますか？ [Y/n/a] : ").strip().lower()

    if ans in ("", "y", "yes"):
        return

    if ans in ("a", "auto"):
        AUTO_MODE = True
        print("以後、自動実行します。")
        return

    if ans in ("n", "no"):
        print("\nシミュレーションを終了します。")
        sys.exit(0)

    print("入力が認識できません。続行します。")


def line(title=""):
    print("\n" + "=" * 70)
    if title:
        print(title)
        print("=" * 70)


def bits_to_str(bits):
    return "".join(str(b) for b in bits)


def xor_str(values):
    return " XOR ".join(str(v) for v in values)


def polynomial_to_string(exponents):
    terms = []

    for e in sorted(exponents, reverse=True):
        if e == 0:
            terms.append("1")
        elif e == 1:
            terms.append("x")
        else:
            terms.append(f"x^{e}")

    return " + ".join(terms)


def choose_random_primitive_polynomial():
    degree = random.choice(list(PRIMITIVE_POLYNOMIALS.keys()))
    exponents = random.choice(PRIMITIVE_POLYNOMIALS[degree])
    return degree, exponents


def polynomial_to_taps(degree, exponents):
    taps = []

    for e in exponents:
        if e == degree:
            continue
        if e == 0:
            taps.append(degree - 1)
        else:
            taps.append(degree - 1 - e)

    return sorted(set(taps))


def random_seed_bits(degree):
    while True:
        seed = [random.randint(0, 1) for _ in range(degree)]
        if any(seed):
            return seed


def exponents_to_generator(exponents):
    degree = max(exponents)
    generator = [0] * (degree + 1)

    for e in exponents:
        generator[degree - e] = 1

    return generator


def crc_polynomial_string(generator):
    degree = len(generator) - 1
    exponents = []

    for i, bit in enumerate(generator):
        if bit == 1:
            exponents.append(degree - i)

    return polynomial_to_string(exponents)


# ------------------------------------------------------------
# Matplotlib 図示
# ------------------------------------------------------------

def draw_register(
    reg,
    title="レジスタ",
    taps=None,
    highlight_index=None,
    max_show=64,
):
    if not SHOW_PLOTS:
        return

    n = len(reg)

    if n > max_show:
        shown = reg[:max_show]
        shown_n = max_show
        note = f"先頭 {max_show} bit のみ表示 / 全体 {n} bit"
    else:
        shown = reg
        shown_n = n
        note = f"{n} bit"

    fig_width = max(8, shown_n * 0.45)
    fig, ax = plt.subplots(figsize=(fig_width, 2.8))

    for i, bit in enumerate(shown):
        rect = Rectangle(
            (i, 0),
            1,
            1,
            fill=False,
            linewidth=2
        )
        ax.add_patch(rect)

        reg_name = f"R{n - 1 - i}"

        ax.text(
            i + 0.5,
            0.72,
            reg_name,
            ha="center",
            va="center",
            fontsize=9
        )

        ax.text(
            i + 0.5,
            0.32,
            str(bit),
            ha="center",
            va="center",
            fontsize=16,
            fontweight="bold"
        )

        if taps and i in taps:
            ax.plot(i + 0.5, 1.12, marker="o", markersize=8)
            ax.text(
                i + 0.5,
                1.32,
                "tap",
                ha="center",
                fontsize=8
            )

        if highlight_index is not None and i == highlight_index:
            ax.plot(i + 0.5, -0.15, marker="v", markersize=10)
            ax.text(
                i + 0.5,
                -0.35,
                "error",
                ha="center",
                fontsize=8
            )

    ax.set_xlim(0, shown_n)
    ax.set_ylim(-0.55, 1.55)
    ax.axis("off")
    ax.set_title(f"{title}  ({note})")

    plt.tight_layout()
    plt.show()


def draw_hamming_code(code, title="ハミング符号", highlight_index=None):
    labels = ["P1", "P2", "D1", "P3", "D2", "D3", "D4"]

    fig, ax = plt.subplots(figsize=(9, 2.6))

    for i, bit in enumerate(code):
        rect = Rectangle(
            (i, 0),
            1,
            1,
            fill=False,
            linewidth=2
        )
        ax.add_patch(rect)

        ax.text(
            i + 0.5,
            0.75,
            labels[i],
            ha="center",
            va="center",
            fontsize=11
        )

        ax.text(
            i + 0.5,
            0.30,
            str(bit),
            ha="center",
            va="center",
            fontsize=18,
            fontweight="bold"
        )

        ax.text(
            i + 0.5,
            -0.25,
            f"{i + 1}",
            ha="center",
            fontsize=10
        )

        if highlight_index is not None and i == highlight_index:
            ax.plot(i + 0.5, 1.2, marker="o", markersize=12)
            ax.text(
                i + 0.5,
                1.4,
                "誤り",
                ha="center",
                fontsize=10
            )

    ax.set_xlim(0, 7)
    ax.set_ylim(-0.5, 1.65)
    ax.axis("off")
    ax.set_title(title)

    plt.tight_layout()
    plt.show()


# ------------------------------------------------------------
# テキスト版レジスタ表示
# ------------------------------------------------------------

def print_register(reg, label="レジスタ", max_show=32):
    n = len(reg)

    print(label)

    if n > max_show:
        print(f"レジスタ長 {n} bit のため、先頭 {max_show} bit のみ表示します。")
        shown = reg[:max_show]
    else:
        shown = reg

    m = len(shown)

    print("+----" * m + "+")
    print("".join(f"|R{n-1-i:<3}" for i in range(m)) + "|")
    print("+----" * m + "+")
    print("".join(f"| {b:<3}" for b in shown) + "|")
    print("+----" * m + "+")

    if n > max_show:
        print("...")


# ------------------------------------------------------------
# 基本シフトレジスタ
# ------------------------------------------------------------

class ShiftRegister:
    def __init__(self, size, init=None):
        self.reg = init[:] if init else [0] * size

    def shift_right_verbose(self, input_bit, clock):
        before = self.reg[:]
        output_bit = self.reg[-1]
        self.reg = [input_bit] + self.reg[:-1]

        line(f"基本シフトレジスタ クロック {clock}")

        print(f"入力ビット : {input_bit}")
        print(f"出力ビット : {output_bit}")

        print()
        print_register(before, "シフト前レジスタ")
        draw_register(before, f"シフト前レジスタ clock={clock}")

        print()
        print_register(self.reg, "シフト後レジスタ")
        draw_register(self.reg, f"シフト後レジスタ clock={clock}")

        wait_next()
        return output_bit


# ------------------------------------------------------------
# LFSR
# ------------------------------------------------------------

class LFSR:
    def __init__(self, seed, taps, exponents):
        if all(b == 0 for b in seed):
            raise ValueError("LFSRの初期値をすべて0にはできません。")

        self.reg = seed[:]
        self.taps = taps[:]
        self.exponents = exponents[:]

    def step_verbose(self, clock):
        before = self.reg[:]
        tap_values = [before[t] for t in self.taps]

        feedback = 0
        for v in tap_values:
            feedback ^= v

        output_bit = before[-1]
        self.reg = [feedback] + before[:-1]

        line(f"LFSR クロック {clock}")

        print("利用している規約多項式")
        print(f"G(x) = {polynomial_to_string(self.exponents)}")

        print()
        print("レジスタ長")
        print(f"{len(before)} bit")

        print()
        print("帰還タップ")
        print("tap positions =", self.taps)
        print("tap values    =", tap_values)

        print()
        print_register(before, "シフト前レジスタ")
        draw_register(
            before,
            title=f"LFSR シフト前 clock={clock}",
            taps=self.taps
        )

        print()
        print("フィードバック計算")
        print(f"feedback = {xor_str(tap_values)} = {feedback}")

        print()
        print(f"出力ビット = {output_bit}")

        print()
        print_register(self.reg, "シフト後レジスタ")
        draw_register(
            self.reg,
            title=f"LFSR シフト後 clock={clock}",
            taps=self.taps
        )

        wait_next()
        return output_bit


# ------------------------------------------------------------
# CRC
# ------------------------------------------------------------

def crc_verbose(data, generator):
    line("CRC計算")

    work = data[:] + [0] * (len(generator) - 1)

    print("入力データ")
    print(bits_to_str(data))

    print()
    print("生成多項式")
    print(f"G(x) = {crc_polynomial_string(generator)}")

    print()
    print("生成多項式ビット列")
    print(bits_to_str(generator))

    print()
    print("初期状態")
    print(bits_to_str(work))
    draw_register(work, "CRC 初期状態")

    wait_next()

    for i in range(len(data)):
        line(f"CRC ステップ {i + 1}")

        print("現在の作業ビット列")
        print(bits_to_str(work))
        draw_register(work, f"CRC ステップ {i + 1} XOR前", highlight_index=i)

        if work[i] == 1:
            print()
            print(f"左から {i + 1} 番目のビットが1なので、生成多項式をXORします。")

            before = work[:]

            for j in range(len(generator)):
                work[i + j] ^= generator[j]

            print()
            print("XOR前")
            print(bits_to_str(before))

            print("XORする位置")
            print(" " * i + bits_to_str(generator))

            print("XOR後")
            print(bits_to_str(work))

            draw_register(work, f"CRC ステップ {i + 1} XOR後")

        else:
            print()
            print(f"左から {i + 1} 番目のビットが0なので、何もしません。")

        wait_next()

    remainder = work[-(len(generator) - 1):]
    codeword = data + remainder

    line("CRC結果")

    print("余り")
    print(bits_to_str(remainder))
    draw_register(remainder, "CRC 余り")

    print()
    print("CRC付き符号語")
    print(bits_to_str(codeword))
    draw_register(codeword, "CRC付き符号語")

    wait_next()

    return codeword, remainder


def crc_check_verbose(codeword, generator):
    line("CRC検査")

    work = codeword[:]

    print("受信語")
    print(bits_to_str(codeword))
    draw_register(work, "CRC検査 受信語")

    print()
    print("生成多項式")
    print(f"G(x) = {crc_polynomial_string(generator)}")

    wait_next()

    for i in range(len(codeword) - len(generator) + 1):
        if work[i] == 1:
            before = work[:]

            for j in range(len(generator)):
                work[i + j] ^= generator[j]

            line(f"CRC検査ステップ {i + 1}")
            print("XOR前 :", bits_to_str(before))
            print("XOR後 :", bits_to_str(work))

            draw_register(before, f"CRC検査 ステップ {i + 1} XOR前", highlight_index=i)
            draw_register(work, f"CRC検査 ステップ {i + 1} XOR後")

            wait_next()

    remainder = work[-(len(generator) - 1):]

    line("CRC検査結果")

    print("検査余り")
    print(bits_to_str(remainder))
    draw_register(remainder, "CRC検査 余り")

    if any(remainder):
        print("判定 : 誤りあり")
    else:
        print("判定 : 誤りなし")

    wait_next()

    return remainder


# ------------------------------------------------------------
# ハミング符号
# ------------------------------------------------------------

def hamming74_encode_verbose(data):
    line("ハミング(7,4) 符号化")

    d1, d2, d3, d4 = data

    print("入力データ")
    print("D1 D2 D3 D4")
    print(f"{d1}  {d2}  {d3}  {d4}")

    p1_values = [d1, d2, d4]
    p2_values = [d1, d3, d4]
    p3_values = [d2, d3, d4]

    p1 = d1 ^ d2 ^ d4
    p2 = d1 ^ d3 ^ d4
    p3 = d2 ^ d3 ^ d4

    print()
    print("パリティビット計算")
    print(f"P1 = D1 XOR D2 XOR D4 = {xor_str(p1_values)} = {p1}")
    print(f"P2 = D1 XOR D3 XOR D4 = {xor_str(p2_values)} = {p2}")
    print(f"P3 = D2 XOR D3 XOR D4 = {xor_str(p3_values)} = {p3}")

    code = [p1, p2, d1, p3, d2, d3, d4]

    print()
    print("配置")
    print("位置 : 1  2  3  4  5  6  7")
    print("内容 : P1 P2 D1 P3 D2 D3 D4")
    print("符号 :", "  ".join(str(b) for b in code))

    print()
    print("生成された符号語")
    print(bits_to_str(code))

    draw_hamming_code(code, "ハミング(7,4) 符号語")

    wait_next()

    return code


def hamming74_decode_verbose(received):
    line("ハミング(7,4) 復号")

    b = received[:]

    print("受信語")
    print(bits_to_str(b))
    draw_hamming_code(b, "ハミング 受信語")

    s1_values = [b[0], b[2], b[4], b[6]]
    s2_values = [b[1], b[2], b[5], b[6]]
    s3_values = [b[3], b[4], b[5], b[6]]

    s1 = b[0] ^ b[2] ^ b[4] ^ b[6]
    s2 = b[1] ^ b[2] ^ b[5] ^ b[6]
    s3 = b[3] ^ b[4] ^ b[5] ^ b[6]

    print()
    print("シンドローム計算")
    print(f"S1 = b1 XOR b3 XOR b5 XOR b7 = {xor_str(s1_values)} = {s1}")
    print(f"S2 = b2 XOR b3 XOR b6 XOR b7 = {xor_str(s2_values)} = {s2}")
    print(f"S3 = b4 XOR b5 XOR b6 XOR b7 = {xor_str(s3_values)} = {s3}")

    error_position = s1 + 2 * s2 + 4 * s3

    print()
    print("シンドローム")
    print(f"S = S3 S2 S1 = {s3}{s2}{s1}")
    print(f"誤り位置 = S1 + 2*S2 + 4*S3 = {error_position}")

    if error_position != 0:
        print()
        print(f"判定 : {error_position} ビット目に誤りがあります。")

        before = b[:]
        draw_hamming_code(before, "訂正前", highlight_index=error_position - 1)

        b[error_position - 1] ^= 1

        print()
        print("訂正前")
        print(bits_to_str(before))

        print("訂正後")
        print(bits_to_str(b))

        draw_hamming_code(b, "訂正後")

    else:
        print()
        print("判定 : 誤りはありません。")

    decoded = [b[2], b[4], b[5], b[6]]

    print()
    print("復号データ")
    print(bits_to_str(decoded))

    wait_next()

    return {
        "syndrome": [s3, s2, s1],
        "error_position": error_position,
        "corrected_code": b,
        "decoded_data": decoded,
    }


# ------------------------------------------------------------
# 誤り挿入
# ------------------------------------------------------------

def inject_error(bits, position=None):
    y = bits[:]

    if position is None:
        position = random.randint(0, len(bits) - 1)

    y[position] ^= 1

    return y, position + 1


# ------------------------------------------------------------
# main
# ------------------------------------------------------------

def main():
    line("教育用通信シミュレータ開始")

    print("操作方法")
    print("Enter または y : 次へ進む")
    print("n              : 終了")
    print("a              : 以後、自動実行")
    print()
    print("注意")
    print("各レジスタ表示ごとに Matplotlib の図が表示されます。")
    print("図のウィンドウを閉じると次の処理へ進みます。")

    wait_next()

    # --------------------------------------------------------
    # ランダムLFSR規約多項式
    # --------------------------------------------------------

    degree, lfsr_exponents = choose_random_primitive_polynomial()
    taps = polynomial_to_taps(degree, lfsr_exponents)
    seed = random_seed_bits(degree)

    theoretical_period = 2 ** degree - 1

    line("今回ランダムに選ばれたLFSR規約多項式")

    print("次数")
    print(degree)

    print()
    print("規約多項式")
    print(f"G(x) = {polynomial_to_string(lfsr_exponents)}")

    print()
    print("帰還タップ")
    print(taps)

    print()
    print("初期値")
    print(bits_to_str(seed))
    draw_register(seed, "LFSR 初期値", taps=taps)

    print()
    print("理論周期")
    print(f"2^{degree} - 1 = {theoretical_period}")

    print()
    print("このデモでは先頭10クロックだけ表示します。")

    wait_next()

    # --------------------------------------------------------
    # 1. 基本シフトレジスタ
    # --------------------------------------------------------

    line("1. 基本シフトレジスタ")

    sr = ShiftRegister(size=8)
    input_bits = [1, 0, 1, 1, 0, 1]

    print("入力ビット列")
    print(bits_to_str(input_bits))

    wait_next()

    for clk, bit in enumerate(input_bits, start=1):
        sr.shift_right_verbose(bit, clk)

    # --------------------------------------------------------
    # 2. LFSR
    # --------------------------------------------------------

    line("2. ランダム規約多項式によるLFSR")

    lfsr = LFSR(
        seed=seed,
        taps=taps,
        exponents=lfsr_exponents
    )

    pn_sequence = []

    for clk in range(1, 11):
        pn_sequence.append(lfsr.step_verbose(clk))

    line("LFSR生成系列 先頭10ビット")
    print(bits_to_str(pn_sequence))
    draw_register(pn_sequence, "LFSR生成系列 先頭10ビット")

    wait_next()

    # --------------------------------------------------------
    # 3. CRC
    # --------------------------------------------------------

    line("3. CRC生成と検査")

    crc_degree, crc_exponents = choose_random_primitive_polynomial()

    while crc_degree > 16:
        crc_degree, crc_exponents = choose_random_primitive_polynomial()

    generator = exponents_to_generator(crc_exponents)
    crc_data = [random.randint(0, 1) for _ in range(8)]

    line("今回ランダムに選ばれたCRC生成多項式")

    print("次数")
    print(crc_degree)

    print()
    print("生成多項式")
    print(f"G(x) = {polynomial_to_string(crc_exponents)}")

    print()
    print("生成多項式ビット列")
    print(bits_to_str(generator))
    draw_register(generator, "CRC 生成多項式ビット列")

    print()
    print("入力データ")
    print(bits_to_str(crc_data))
    draw_register(crc_data, "CRC 入力データ")

    wait_next()

    crc_codeword, crc_remainder_bits = crc_verbose(
        data=crc_data,
        generator=generator
    )

    crc_received, crc_error_position = inject_error(
        crc_codeword
    )

    line("CRC用 1ビット誤り挿入")

    print("送信語")
    print(bits_to_str(crc_codeword))

    print()
    print(f"誤り挿入位置 : {crc_error_position} ビット目")

    print()
    print("受信語")
    print(bits_to_str(crc_received))

    draw_register(
        crc_received,
        "CRC 受信語 誤りあり",
        highlight_index=crc_error_position - 1
    )

    wait_next()

    crc_check_verbose(
        codeword=crc_received,
        generator=generator
    )

    # --------------------------------------------------------
    # 4. ハミング符号
    # --------------------------------------------------------

    line("4. ハミング符号")

    data = [random.randint(0, 1) for _ in range(4)]

    code = hamming74_encode_verbose(data)

    received, error_position = inject_error(code)

    line("ハミング符号用 1ビット誤り挿入")

    print("送信語")
    print(bits_to_str(code))
    draw_hamming_code(code, "ハミング 送信語")

    print()
    print(f"誤り挿入位置 : {error_position} ビット目")

    print()
    print("受信語")
    print(bits_to_str(received))
    draw_hamming_code(
        received,
        "ハミング 受信語 誤りあり",
        highlight_index=error_position - 1
    )

    wait_next()

    result = hamming74_decode_verbose(received)

    # --------------------------------------------------------
    # まとめ
    # --------------------------------------------------------

    line("まとめ")

    print("LFSR")
    print(f"次数       : {degree}")
    print(f"規約多項式 : G(x) = {polynomial_to_string(lfsr_exponents)}")
    print(f"理論周期   : 2^{degree} - 1 = {theoretical_period}")
    print(f"生成系列   : {bits_to_str(pn_sequence)}")

    print()
    print("CRC")
    print(f"次数       : {crc_degree}")
    print(f"生成多項式 : G(x) = {polynomial_to_string(crc_exponents)}")
    print(f"入力       : {bits_to_str(crc_data)}")
    print(f"CRC余り    : {bits_to_str(crc_remainder_bits)}")
    print(f"符号語     : {bits_to_str(crc_codeword)}")

    print()
    print("ハミング(7,4)")
    print(f"元データ   : {bits_to_str(data)}")
    print(f"符号語     : {bits_to_str(code)}")
    print(f"受信語     : {bits_to_str(received)}")
    print(f"誤り位置   : {result['error_position']} ビット目")
    print(f"訂正後     : {bits_to_str(result['corrected_code'])}")
    print(f"復号結果   : {bits_to_str(result['decoded_data'])}")

    print()
    print("シミュレーション完了")


if __name__ == "__main__":
    main()
```
