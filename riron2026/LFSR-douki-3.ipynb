```python
import numpy as np
import matplotlib.pyplot as plt

# ============================================================
# LFSR 複数系列・同期補足・相関解析 完全版
# ============================================================

np.random.seed(0)

# ------------------------------------------------------------
# 基本関数
# ------------------------------------------------------------

def lfsr_sequence(seed, taps, n_bits):
    """
    LFSRで0/1系列を生成する。
    seed : 初期状態 例 [1,0,0,0,1]
    taps : XORを取る位置 例 [2,4]
    n_bits : 生成ビット数
    """
    state = seed[:]
    seq = []

    for _ in range(n_bits):
        out = state[-1]
        seq.append(out)

        feedback = 0
        for t in taps:
            feedback ^= state[t]

        state = [feedback] + state[:-1]

    return np.array(seq, dtype=int)


def bits_to_pm1(bits):
    """
    0,1 を +1,-1 に変換する。
    相関計算では ±1 表現の方が分かりやすい。
    """
    return np.where(bits == 0, 1, -1)


def cyclic_shift(seq, shift):
    return np.roll(seq, shift)


def cyclic_corr(a, b):
    """
    a と b の巡回相関を全シフトについて計算する。
    """
    x = bits_to_pm1(a)
    y = bits_to_pm1(b)

    corr = []
    for shift in range(len(a)):
        shifted = np.roll(y, -shift)
        corr.append(np.sum(x * shifted))

    return np.array(corr)


def max_corr_and_shift(a, b):
    corr = cyclic_corr(a, b)
    shift = int(np.argmax(corr))
    value = int(corr[shift])
    return value, shift, corr


def add_bit_errors(seq, error_rate):
    """
    指定確率でビット反転ノイズを加える。
    """
    noisy = seq.copy()
    mask = np.random.rand(len(seq)) < error_rate
    noisy[mask] ^= 1
    return noisy, mask


def random_nonzero_seed(m):
    """
    全ゼロでないランダム初期値を作る。
    """
    while True:
        seed = np.random.randint(0, 2, size=m).tolist()
        if sum(seed) > 0:
            return seed


# ------------------------------------------------------------
# 設定
# ------------------------------------------------------------

m = 5
period = 2**m - 1

# 例：x^5 + x^2 + 1 のようなLFSR例
# tapsはPythonのindexなので0始まり
taps_1 = [2, 4]

# もう一つのLFSR例
taps_2 = [1, 4]

num_sequences = 6
error_rate = 0.10
num_trials = 200

print("====================================")
print("LFSR設定")
print("====================================")
print("段数 m:", m)
print("周期:", period)
print("LFSR1 taps:", taps_1)
print("LFSR2 taps:", taps_2)
print("生成系列数:", num_sequences)
print("雑音率:", error_rate)
print("試行回数:", num_trials)


# ------------------------------------------------------------
# 複数LFSR系列の生成
# ------------------------------------------------------------

seeds = [random_nonzero_seed(m) for _ in range(num_sequences)]
sequences = []

for seed in seeds:
    seq = lfsr_sequence(seed, taps_1, period)
    sequences.append(seq)

print("\n====================================")
print("生成されたLFSR系列")
print("====================================")

for i, (seed, seq) in enumerate(zip(seeds, sequences)):
    print(f"S{i+1} seed={seed} : {''.join(map(str, seq))}")


# ------------------------------------------------------------
# Gold系列の生成
# ------------------------------------------------------------

seed_a = [1, 0, 0, 0, 1]
seed_b = [1, 1, 0, 0, 1]

seq_a = lfsr_sequence(seed_a, taps_1, period)
seq_b = lfsr_sequence(seed_b, taps_2, period)

gold_sequences = []

for shift in range(period):
    gold = seq_a ^ np.roll(seq_b, shift)
    gold_sequences.append(gold)

print("\n====================================")
print("Gold系列の例")
print("====================================")

for i in range(5):
    print(f"G{i+1}: {''.join(map(str, gold_sequences[i]))}")


# ------------------------------------------------------------
# 相関行列の作成
# ------------------------------------------------------------

def correlation_matrix(seq_list):
    n = len(seq_list)
    mat = np.zeros((n, n), dtype=int)
    shift_mat = np.zeros((n, n), dtype=int)

    for i in range(n):
        for j in range(n):
            value, shift, _ = max_corr_and_shift(seq_list[i], seq_list[j])
            mat[i, j] = value
            shift_mat[i, j] = shift

    return mat, shift_mat


lfsr_corr_matrix, lfsr_shift_matrix = correlation_matrix(sequences)

gold_num = 10
gold_corr_matrix, gold_shift_matrix = correlation_matrix(gold_sequences[:gold_num])

print("\n====================================")
print("LFSR系列 最大相関行列")
print("====================================")
print(lfsr_corr_matrix)

print("\n====================================")
print("LFSR系列 最大相関シフト行列")
print("====================================")
print(lfsr_shift_matrix)

print("\n====================================")
print("Gold系列 最大相関行列")
print("====================================")
print(gold_corr_matrix)


# ------------------------------------------------------------
# 同期補足テスト
# ------------------------------------------------------------

target_index = 2
true_shift = 9

reference = sequences[target_index]
received_clean = cyclic_shift(reference, true_shift)
received_noisy, error_mask = add_bit_errors(received_clean, error_rate)

corr_clean = cyclic_corr(reference, received_clean)
corr_noisy = cyclic_corr(reference, received_noisy)

estimated_shift_clean = int(np.argmax(corr_clean))
estimated_shift_noisy = int(np.argmax(corr_noisy))

print("\n====================================")
print("同期補足テスト")
print("====================================")
print("対象系列:", f"S{target_index+1}")
print("本当のシフト:", true_shift)
print("推定シフト 雑音なし:", estimated_shift_clean)
print("推定シフト 雑音あり:", estimated_shift_noisy)
print("最大相関 雑音なし:", corr_clean[estimated_shift_clean])
print("最大相関 雑音あり:", corr_noisy[estimated_shift_noisy])
print("ビット誤り数:", int(error_mask.sum()))


# ------------------------------------------------------------
# 複数回試行による同期成功率
# ------------------------------------------------------------

success_count = 0
peak_values = []
estimated_shifts = []
true_shifts = []

for _ in range(num_trials):
    true_s = np.random.randint(0, period)
    received = cyclic_shift(reference, true_s)
    noisy, _ = add_bit_errors(received, error_rate)

    corr = cyclic_corr(reference, noisy)
    est_s = int(np.argmax(corr))
    peak = int(corr[est_s])

    if est_s == true_s:
        success_count += 1

    true_shifts.append(true_s)
    estimated_shifts.append(est_s)
    peak_values.append(peak)

success_rate = success_count / num_trials

print("\n====================================")
print("同期補足 統計")
print("====================================")
print("試行回数:", num_trials)
print("成功回数:", success_count)
print("成功率:", round(success_rate * 100, 2), "%")
print("平均最大相関:", round(float(np.mean(peak_values)), 2))
print("最小最大相関:", int(np.min(peak_values)))
print("最大最大相関:", int(np.max(peak_values)))


# ------------------------------------------------------------
# プロット1：複数系列のデジタル波形
# ------------------------------------------------------------

plt.figure(figsize=(12, 5))

for i, seq in enumerate(sequences):
    y = bits_to_pm1(seq) + i * 2.5
    plt.step(range(period), y, where="post", label=f"S{i+1}")

plt.title("生成された複数LFSR系列")
plt.xlabel("bit position")
plt.ylabel("sequence")
plt.yticks([])
plt.grid(True)
plt.legend(loc="upper right")
plt.tight_layout()
plt.show()


# ------------------------------------------------------------
# プロット2：LFSR系列の最大相関ヒートマップ
# ------------------------------------------------------------

plt.figure(figsize=(7, 6))
plt.imshow(lfsr_corr_matrix)
plt.colorbar(label="max correlation")

plt.xticks(range(num_sequences), [f"S{i+1}" for i in range(num_sequences)])
plt.yticks(range(num_sequences), [f"S{i+1}" for i in range(num_sequences)])

for i in range(num_sequences):
    for j in range(num_sequences):
        plt.text(j, i, str(lfsr_corr_matrix[i, j]),
                 ha="center", va="center")

plt.title("LFSR系列の最大相関行列")
plt.tight_layout()
plt.show()


# ------------------------------------------------------------
# プロット3：Gold系列の最大相関ヒートマップ
# ------------------------------------------------------------

plt.figure(figsize=(8, 7))
plt.imshow(gold_corr_matrix)
plt.colorbar(label="max correlation")

plt.xticks(range(gold_num), [f"G{i+1}" for i in range(gold_num)])
plt.yticks(range(gold_num), [f"G{i+1}" for i in range(gold_num)])

for i in range(gold_num):
    for j in range(gold_num):
        plt.text(j, i, str(gold_corr_matrix[i, j]),
                 ha="center", va="center")

plt.title("Gold系列の最大相関行列")
plt.tight_layout()
plt.show()


# ------------------------------------------------------------
# プロット4：同期補足の相関グラフ
# ------------------------------------------------------------

plt.figure(figsize=(12, 4))
plt.stem(range(period), corr_noisy)

plt.axvline(true_shift, linestyle="--", label=f"true shift = {true_shift}")
plt.axvline(estimated_shift_noisy, linestyle=":", label=f"estimated shift = {estimated_shift_noisy}")

plt.title("同期補足：雑音あり受信系列との全位相相関")
plt.xlabel("shift")
plt.ylabel("correlation")
plt.grid(True)
plt.legend()
plt.tight_layout()
plt.show()


# ------------------------------------------------------------
# プロット5：雑音なし vs 雑音ありの比較
# ------------------------------------------------------------

plt.figure(figsize=(12, 4))

plt.plot(corr_clean, marker="o", label="clean")
plt.plot(corr_noisy, marker="x", label="noisy")

plt.axvline(true_shift, linestyle="--", label=f"true shift = {true_shift}")

plt.title("相関ピークの比較：雑音なし vs 雑音あり")
plt.xlabel("shift")
plt.ylabel("correlation")
plt.grid(True)
plt.legend()
plt.tight_layout()
plt.show()


# ------------------------------------------------------------
# プロット6：自己相関と相互相関の比較
# ------------------------------------------------------------

auto_corr = cyclic_corr(sequences[0], sequences[0])
cross_corr = cyclic_corr(sequences[0], sequences[1])

plt.figure(figsize=(12, 4))
plt.plot(auto_corr, marker="o", label="S1 vs S1 self-correlation")
plt.plot(cross_corr, marker="x", label="S1 vs S2 cross-correlation")

plt.title("自己相関と相互相関の比較")
plt.xlabel("shift")
plt.ylabel("correlation")
plt.grid(True)
plt.legend()
plt.tight_layout()
plt.show()


# ------------------------------------------------------------
# プロット7：同期補足の成功・失敗分布
# ------------------------------------------------------------

correct = np.array(true_shifts) == np.array(estimated_shifts)

plt.figure(figsize=(10, 4))
plt.hist(peak_values, bins=15)

plt.title("同期補足で得られた最大相関値の分布")
plt.xlabel("maximum correlation")
plt.ylabel("count")
plt.grid(True)
plt.tight_layout()
plt.show()


# ------------------------------------------------------------
# プロット8：真のシフトと推定シフト
# ------------------------------------------------------------

plt.figure(figsize=(6, 6))
plt.scatter(true_shifts, estimated_shifts)

plt.plot([0, period - 1], [0, period - 1], linestyle="--")

plt.title("真のシフトと推定シフト")
plt.xlabel("true shift")
plt.ylabel("estimated shift")
plt.grid(True)
plt.tight_layout()
plt.show()


# ------------------------------------------------------------
# 最後のまとめ
# ------------------------------------------------------------

print("\n====================================")
print("まとめ")
print("====================================")
print("1. 同じLFSRで初期値だけ変えた系列は、位相違いになりやすいです。")
print("2. 自己相関は正しい位相で大きなピークを持ちます。")
print("3. 雑音が入るとピークは下がりますが、最大位置が残れば同期できます。")
print("4. Gold系列は複数ユーザー用の系列比較に向いています。")
print("5. 成功率を見ると、雑音率に対する同期補足の強さを評価できます。")
```
