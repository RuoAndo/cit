```python
import numpy as np
import matplotlib.pyplot as plt

P0 = np.arange(0,1, 0.01) # レベル0の出現確率P0
P1 = 1 - P0 # レベル1の出現確率P1

# エントロピーの算出
H = -P0 * np.log2(P0) -P1 * np.log2(P1)

# グラフ描画設定
plt.plot(P0, H)
plt.xlabel("P0")
plt.ylabel("H")
plt.title('Entropy')

# ラベルの描画
plt.legend()
# グラフの描画実行
plt.show()
```

```python
import numpy as np
import matplotlib.pyplot as plt

SN= np.arange(0,1, 0.01) 
B = 1

# エントロピーの算出
C = B * np.log2(SN)

# グラフ描画設定
plt.plot(SN, C)
plt.xlabel("Signal-Noise Ratio")
plt.ylabel("Capacity")
plt.title('Shannon Formula')

# ラベルの描画
plt.legend()
# グラフの描画実行
plt.show()
```

```python
import cv2
import matplotlib.pyplot as plt
import numpy as np

img = cv2.imread('./lena.png') #ファイルのバスは適宜変えてください
height, width, _ = img.shape

# ヒストグラム（各色の画素数）の算出
histgram = [0]*256
for i in range(height):
    for j in range(width):
        histgram[img[i, j, 0]] += 1

# エントロピーの算出
size = height * width
entropy = 0

for i in range(256):
    # レベルiの出現確率p
    p = histgram[i]/size
    if p == 0:
        continue
    entropy -= p*np.log2(p)

plt.imshow(img)
print('Lena {}:'.format(entropy))
```
